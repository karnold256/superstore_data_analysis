```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import plot

#reading in the dataset
data = pd.read_excel('C:\PowerBI Data Sets\superstore.xlsx')
data

#checking the data set for null values
data.isnull()

#checking to get the columns in the data set
data.columns

#Row ID', 'Order Priority', 'Discount', 'Unit Price', 'Shipping Cost',
       #'Customer ID', 'Customer Name', 'Ship Mode', 'Customer Segment',
       #'Product Category', 'Product Sub-Category', 'Product Container',
       #'Product Name', 'Product Base Margin', 'Region', 'State or Province',
       #'City', 'Postal Code', 'Order Date', 'Ship Date', 'Profit',
       #'Quantity ordered new', 'Sales', 'Order ID'


#getting the customer's name without having them repeated
customer  = data['Customer Name'].unique()
customer

#this describes the sample data
data.describe

highest_profit = data['Profit'].max()
highest_profit

lowest_profit = data['Profit'].min()
lowest_profit

#counting the number of customers
customers = data['Customer Name'].unique()
customers
num_of_customers = data['Sales'].value_counts()
num_of_customers

#the nornalize creates a percentage
num_of_state = data['State or Province'].value_counts(normalize=True)
num_of_state

#this is counting the number of orders in each state
num_of_state = data['State or Province'].value_counts()
num_of_state

#we cann use group by function
state_grouped = data.groupby(['State or Province'])
#grouped data by state and started with missouri state
Missouri = state_grouped.get_group('Missouri')
Missouri.to_excel("Missouri.xlsx")

#we then grouped and isolated kansas state
Kansas = state_grouped.get_group('Kansas')
Kansas.to_excel("Kansas.xlsx")

#we can get the median sales from each state
median_state_sales = state_grouped['Sales'].median()
median_state_sales

#we can then extract the lowest median sales made and the highest median sales made
lowest_median_sales_made = median_state_sales.min()
lowest_median_sales_made

#we then get the state with the lowest median sales
#state_with_lowest_median_sales = median_state_sales.min().loc['State or Province']
#state_with_lowest_median_sales
highest_median_sales_made = median_state_sales.max()
highest_median_sales_made

#filtering using to get the any results getting different product categories ordered in california
filtered = data['State or Province'] == 'California'
category = data.loc[filtered]['Product Category']
category

#getting the total of all the products categories and then it gives the sum of all products that are office supplies
filtered = data['State or Province'] == 'Maryland'
category = data.loc[filtered]['Product Category'].str.contains('Office Supplies').sum()
category

#sum of orders of office supplies from every state
total_of_states = state_grouped['Product Category'].apply(lambda  x : x.str.contains('Office Supplies').sum())
total_of_states

#sum of orders of office supplies from every state
total_of_office_supplies_in_states = state_grouped['Product Category'].apply(lambda  x : x.str.contains('Office Supplies').sum())
total_of_office_supplies_in_states

#renaming the office supplies column
#office_supplies_renamed = total_of_office_supplies_in_states.rename(columns=({'State or Province': 'State', 'Product Category': 'Office Supplies'}), inplace=True)
#office_supplies_renamed

#total of technology sales in each state
#sum of orders of technology from every state
total_of_technology_in_states = state_grouped['Product Category'].apply(lambda  x : x.str.contains('Technology').sum())
total_of_technology_in_states

#sum of orders of furniture from every state
total_of_furniture_in_states = state_grouped['Product Category'].apply(lambda  x : x.str.contains('Furniture').sum())
total_of_furniture_in_states

#combining the three data frames into one we use the concatonate method in pandas
total_of_category_by_state = pd.concat([total_of_office_supplies_in_states, total_of_furniture_in_states, total_of_technology_in_states], axis='columns', sort=False)
total_of_category_by_state
total_of_category_by_state.to_excel('Category Totals by States.xlsx')

#remaning the headers to have a useful meaning or names
#total_of_category_by_state_rename = total_of_category_by_state.rename(columns={'State or Province': 'State', 'Product Category': 'Office Supplies', 'Product Category': 'Technology', 'Product Category': 'Furniture'})
#total_of_category_by_state_rename

#aggregated info for a particular state missouri
aggregated_info_per_state_2 = state_grouped['Sales'].agg(['median', 'sum', 'mean', 'min', 'max']).loc['Missouri']
aggregated_info_per_state_2

#using the aggregated function we can find
aggregated_info_per_state = state_grouped['Sales'].agg(['median', 'sum', 'mean', 'min', 'max'])
aggregated_info_per_state
aggregated_info_per_state.to_excel('Aggregated Functions Per State.xlsx')

percentage_of_profit = (data['Profit'] / data['Sales'] * 100)
percentage_of_profit

#calculating the percentage of project for every state
profit_in_percentage =pd.concat([data['State or Province'], percentage_of_profit], axis='columns', sort=False).sum()
profit_in_percentage
updated_percent = profit_in_percentage.rename(columns={'State or Province':'State',0: 'Percentage of Profit'})
#updated_percent.to_excel('Profit in Percentage.xlsx')
updated_percent

```
